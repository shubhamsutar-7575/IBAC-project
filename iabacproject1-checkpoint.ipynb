## Domain Analysis

EmpNumber: Unique employee ID (to be removed in future).
Age: Employee's age.
Gender: Male/Female.
MaritalStatus: Single/Married/Divorced.
EducationBackground: Employee's field of education.
EmpDepartment: Department where the employee works (e.g., Sales, HR).
EmpJobRole: Specific job role (e.g., Manager, Sales Executive).
EmpJobLevel: Job seniority level (1 to 5).
BusinessTravelFrequency: Frequency of business travel (Rarely/Frequently/Never).
DistanceFromHome: Distance to the workplace.
EmpEnvironmentSatisfaction: Work environment satisfaction (1 to 4).
EmpJobSatisfaction: Satisfaction with job role (1 to 4).
EmpRelationshipSatisfaction: Satisfaction with work relationships (1 to 4).
EmpJobInvolvement: Job involvement level (1 to 4).
EmpWorkLifeBalance: Work-life balance (1 to 4).
EmpHourlyRate: Hourly pay rate.
EmpLastSalaryHikePercent: Percentage of last salary hike.
TotalWorkExperienceInYears: Total years of professional work.
ExperienceYearsAtThisCompany: Total years in the current company.
ExperienceYearsInCurrentRole: Total years in the current role.
YearsSinceLastPromotion: Years since the last promotion.
YearsWithCurrManager: Years under the current manager.
NumCompaniesWorked: Number of companies the employee has worked for.
PerformanceRating: Employee's performance rating (1 to 4).
TrainingTimesLastYear: Number of trainings attended last year.
OverTime: Indicates if the employee works overtime (Yes/No).
Attrition: Indicates if the employee left the organization (Yes/No).
